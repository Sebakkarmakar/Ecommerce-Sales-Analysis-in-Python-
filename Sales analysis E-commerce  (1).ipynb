```python
df['month']=pd.to_datetime(df['Order Date'])
df['month_name']=df['month'].dt.month_name()
print(df)
```

```python
df['total_sales']=(df['Sales']*df['Quantity']).round(0).astype(int)
print(df)
```

```python
monthly_sales=df.groupby('month_name')[['total_sales']].sum()
print(monthly_sales)
```

```python
maxi_sales_month=df['total_sales'].max()
print(maxi_sales_month)
```

```python
plt.figure(figsize=(15,6))
plt.plot(monthly_sales.index,monthly_sales.values,linestyle="-",marker='o')
plt.xlabel("Month",fontweight='bold')
plt.ylabel("Total_Sales",fontweight="bold")
plt.show()
```

```python
 In this Line plot we can see that the highest peak of the sales in November and Lowest Peak in February .
SO highest total sales  in November month and the Lowest sales in February month.
```

```python
category_wise_sales=df.groupby('Category')[['total_sales']].sum()
print(category_wise_sales)
```

```python
plt.figure(figsize=(8, 8)) # Set a good figure size for a pie chart

category_wise_sales['total_sales'].plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.show()
```
